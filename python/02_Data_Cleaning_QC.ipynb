```python
# Versioned Clean Export and Quality Check Summary
import pandas as pd
import numpy as np

raw=pd.read_excel("Enterprise_Fraud_Transaction_Dataset.xlsx",sheet_name="Transactions")

df=raw.copy()
df["transaction_datetime"]=pd.to_datetime(df["transaction_timestamp"],errors="coerce")

invalid_ts=int(df["transaction_datetime"].isna().sum())
invalid_amt=int((df["transaction_amount"].isna() | (df["transaction_amount"]<=0)).sum())
dup_txn=int(df["transaction_id"].duplicated().sum())

df=df[df["transaction_datetime"].notna()]
df=df[df["transaction_amount"].notna()]
df=df[df["transaction_amount"]>0]

df["customer_age"]=df["customer_age"].where(df["customer_age"].between(18,100),np.nan)

df=df.sort_values("transaction_datetime")
df=df.drop_duplicates(subset=["transaction_id"],keep="last")

qc={
 "raw_rows":int(raw.shape[0]),
 "invalid_timestamp_rows":invalid_ts,
 "invalid_amount_rows":invalid_amt,
 "duplicate_transaction_id_rows":dup_txn,
 "clean_rows":int(df.shape[0])
}
print(qc)

out_name="clean_transactions_v1.csv"
df.to_csv(out_name,index=False)
print(out_name)
```
