```python
import pandas as pd
import numpy as np
```

```python
path=r"Enterprise_Fraud_Transaction_Dataset.xlsx"
df=pd.read_excel(path, sheet_name="Transactions")

print("rows,cols:",df.shape)
print("columns:",list(df.columns))

fraud_counts=df["fraud_label"].value_counts(dropna=False)
print("fraud_counts:",fraud_counts.to_dict())
print("fraud_pct:",(fraud_counts/len(df)*100).round(2).to_dict())

print("top_nulls:",df.isna().sum().sort_values(ascending=False).head(10).to_dict())

print("amount_summary:",df["transaction_amount"].describe(percentiles=[0.95,0.99]).to_dict())

top=df.sort_values("transaction_amount", ascending=False).head(10)[["transaction_id","transaction_amount","fraud_label","channel","customer_region","merchant_category","previous_fraud_flag","is_international"]]
print(top.to_string(index=False))
```

```python
import pandas as pd
import numpy as np

df=pd.read_excel("Enterprise_Fraud_Transaction_Dataset.xlsx",sheet_name="Transactions")

print("shape:",df.shape)
print("fraud_rate_pct:",round(df["fraud_label"].mean()*100,2))
print("fraud_loss_exposure:",round(df.loc[df["fraud_label"]==1,"transaction_amount"].sum(),2))

print("missing_top:",df.isna().sum().sort_values(ascending=False).head(10).to_dict())

p95=np.percentile(df["transaction_amount"],95)
p99=np.percentile(df["transaction_amount"],99)
mx=df["transaction_amount"].max()
print("p95:",round(p95,2),"p99:",round(p99,2),"max:",round(mx,2))

top=df.sort_values("transaction_amount",ascending=False).head(10)[["transaction_id","transaction_amount","fraud_label","channel","customer_region","merchant_category","is_international","previous_fraud_flag"]]
print(top.to_string(index=False))
```
